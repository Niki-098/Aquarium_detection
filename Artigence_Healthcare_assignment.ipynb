```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # Loading the YOLOv8 nano model

# Training the model on CPU
model.train(data='/content/drive/MyDrive/Assignment/aquarium.v2-release.yolov8/data.yaml', epochs=50, imgsz=640, batch=16, device='cpu',project='/content/drive/MyDrive/Assignment/training_results')
```

```python
metrics = model.val()
print(metrics)
```

```python
# Running validation to compute mAP on the validation dataset
metrics = model.val(
    data='/content/drive/MyDrive/Assignment/aquarium.v2-release.yolov8/data.yaml',
    batch=16,
    imgsz=640
)
```

```python
# It can be replaced 'yolov8n' with your model checkpoint
!yolo val model=/content/yolov8n.pt data=/content/drive/MyDrive/Assignment/aquarium.v2-release.yolov8/data.yaml imgsz=640
```

```python
# Performing prediction on the image
results = model.predict('/content/drive/MyDrive/Assignment/aquarium.v2-release.yolov8/test/images/IMG_2514_jpeg_jpg.rf.2826628a3905e2349c7731e1b7a57724.jpg', conf=0.5)


# Loop through the results and display each one
for result in results:
    # Display the image with predictions
    annotated_img = result.plot()  # It returns an annotated image as a numpy array
    # Show the image in notebook
    from matplotlib import pyplot as plt
    plt.imshow(annotated_img)
    plt.axis('off')
    plt.show()
```
